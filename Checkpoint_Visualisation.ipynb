```python
import plotly.express as px
import pandas as pd
```

```python
data = pd.read_csv("Africa_climate_change.csv")
data
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
column_names = data.columns
print(column_names)
```

```python
# Tracez un graphique linéaire pour montrer les fluctuations de température moyennes en Tunisie et au Cameroun.
# Interprétez les résultats.
data['DATE'] = data['DATE'].astype(str)
data['Year'] = data['DATE'].str[:4].astype(int)

# Filtrer les données entre 1980 et 2005 en utilisant la colonne 'Year'
filtered_data = data[(data['Year'] >= 1980) & (data['Year'] <= 2005)]

tunisia_data = data[data['COUNTRY'] == 'Tunisia']
cameroon_data = data[data['COUNTRY'] == 'Cameroon']

# Créer un graphique linéaire.
fig = px.line(data_frame=tunisia_data, x='DATE', y='TAVG', labels={'TAVG': 'Température (°C)'}, title='Température Tunisie (1980-2005)')
fig.add_scatter(x=cameroon_data['DATE'], y=cameroon_data['TAVG'], mode='lines', name='Température au Cameroun')
fig.update_xaxes(title_text='Année')
fig.show()
```

```python
# Créez des histogrammes pour montrer la répartition de la température au Sénégal entre [1980,2000] et [2000,2023] 
# (dans la même figure). Décrivez les résultats obtenus.
```

```python

# Filtrer les données pour chaque période
senegal_1980_2000 = data[(data['COUNTRY'] == 'Senegal') & (data['Year'] >= 1980) & (data['Year'] <= 2000)]
senegal_2000_2023 = data[(data['COUNTRY'] == 'Senegal') & (data['Year'] > 2000)]

# Créer un histogramme pour la période 1980-2000
fig1 = px.histogram(senegal_1980_2000, x='TAVG', title='1980-2000')
fig1.update_xaxes(title_text='Température (°C)')

# Créer un histogramme pour la période 2000-2023
fig2 = px.histogram(senegal_2000_2023, x='TAVG', title='2000-2023')
fig2.update_xaxes(title_text='Température (°C')

# Disposer les deux graphiques côte à côte
fig1.update_layout(showlegend=False)
fig2.update_layout(showlegend=False)

fig1.show()
fig2.show()
```

```python
# Sélectionnez le meilleur graphique pour afficher la température moyenne par pays.
```

```python

# Groupons les données par pays et calculons la température moyenne pour chaque pays.
average_temp_by_country = data.groupby('COUNTRY')['TAVG'].mean().reset_index()

# Créer un graphique en barres interactif.
fig = px.bar(average_temp_by_country, x='COUNTRY', y='TAVG',color='COUNTRY', labels={'TAVG': 'Température Moyenne (°C)'}, title='Température Moyenne par Pays (1980-2005)')
fig.update_xaxes(title_text='Pays')
fig.show()
```

### Comment la température moyenne a-t-elle évolué au fil du temps dans l'ensemble des pays ?
Pour reprondre a cette question , nous allons creer un graphique lineaire interactif pour suivre l'évolution de la température moyenne au fil des années.

```python
fig = px.line(data, x='DATE', y='TAVG', color='COUNTRY', labels={'TAVG': 'Température Moyenne (°C)', 'DATE': 'Année'}, title='Évolution de la Température Moyenne par Pays (1980-2005)')
fig.update_xaxes(title_text='Année')
fig.update_yaxes(title_text='Température Moyenne (°C)')
fig.show()
```
