```python
import plotly.express as px
import pandas as pd
```

```python
data.dropna(inplace=True)
```

```python
column_names = data.columns
print(column_names)
```

```python
# Tracez un graphique linÃ©aire pour montrer les fluctuations de tempÃ©rature moyennes en Tunisie et au Cameroun.
# InterprÃ©tez les rÃ©sultats.
data['DATE'] = data['DATE'].astype(str)
data['Year'] = data['DATE'].str[:4].astype(int)

# Filtrer les donnÃ©es entre 1980 et 2005 en utilisant la colonne 'Year'
filtered_data = data[(data['Year'] >= 1980) & (data['Year'] <= 2005)]

tunisia_data = data[data['COUNTRY'] == 'Tunisia']
cameroon_data = data[data['COUNTRY'] == 'Cameroon']

# CrÃ©er un graphique linÃ©aire.
fig = px.line(data_frame=tunisia_data, x='DATE', y='TAVG', labels={'TAVG': 'TempÃ©rature (Â°C)'}, title='TempÃ©rature Tunisie (1980-2005)')
fig.add_scatter(x=cameroon_data['DATE'], y=cameroon_data['TAVG'], mode='lines', name='TempÃ©rature au Cameroun')
fig.update_xaxes(title_text='AnnÃ©e')
fig.show()
```

```python
# CrÃ©ez des histogrammes pour montrer la rÃ©partition de la tempÃ©rature au SÃ©nÃ©gal entre [1980,2000] et [2000,2023] 
# (dans la mÃªme figure). DÃ©crivez les rÃ©sultats obtenus.
```

```python

# Filtrer les donnÃ©es pour chaque pÃ©riode
senegal_1980_2000 = data[(data['COUNTRY'] == 'Senegal') & (data['Year'] >= 1980) & (data['Year'] <= 2000)]
senegal_2000_2023 = data[(data['COUNTRY'] == 'Senegal') & (data['Year'] > 2000)]

# CrÃ©er un histogramme pour la pÃ©riode 1980-2000
fig1 = px.histogram(senegal_1980_2000, x='TAVG', title='1980-2000')
fig1.update_xaxes(title_text='TempÃ©rature (Â°C)')

# CrÃ©er un histogramme pour la pÃ©riode 2000-2023
fig2 = px.histogram(senegal_2000_2023, x='TAVG', title='2000-2023')
fig2.update_xaxes(title_text='TempÃ©rature (Â°C')

# Disposer les deux graphiques cÃ´te Ã  cÃ´te
fig1.update_layout(showlegend=False)
fig2.update_layout(showlegend=False)

fig1.show()
fig2.show()
```

```python
# SÃ©lectionnez le meilleur graphique pour afficher la tempÃ©rature moyenne par pays.
```

```python

# Groupons les donnÃ©es par pays et calculons la tempÃ©rature moyenne pour chaque pays.
average_temp_by_country = data.groupby('COUNTRY')['TAVG'].mean().reset_index()

# CrÃ©er un graphique en barres interactif.
fig = px.bar(average_temp_by_country, x='COUNTRY', y='TAVG',color='COUNTRY', labels={'TAVG': 'TempÃ©rature Moyenne (Â°C)'}, title='TempÃ©rature Moyenne par Pays (1980-2005)')
fig.update_xaxes(title_text='Pays')
fig.show()
```
